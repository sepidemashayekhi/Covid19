## Prepare Data for Machine learning

```python
data2=pd.read_csv('owid-covid-data.csv')
data2
```

```python
death=data2.loc[:,['location','total_deaths','date','total_cases','new_cases']]
```

```python
death.isnull().sum()
```

```python
death.dropna(inplace=True)
death
```

```python
df1=data[data['country']=='United States']
df1
```

```python
death=death[death['location']=='United States']
death
```

```python
death=death.iloc[320:493,:]
```

```python
death
```

```python
death.isnull().sum()
```

```python
new_death=np.zeros((173,1),dtype='uint16')
death_va=death['total_deaths'].values.astype('uint16')
death_va=np.reshape(death_va,(-1,1))
```

```python
for i in range(0,173):
    if death_va[i]>death_va[i-1]:
        new_death[i]=death_va[i]-death_va[i-1]
    elif  death_va[i]<death_va[i-1]:
         new_death[i]=-(death_va[i-1]-death_va[i])
        
```

```python
df=pd.DataFrame(new_death,columns=['New_death'])
df
```

```python
df['Contry']='USA'
```

```python
date=death['date'].values
df['date']=date
```

```python
total_vaccin=df1['total_vaccinations'].values
df['total_vaccin']=total_vaccin
```

```python
total_cases=death['total_cases'].values
df['total_cases']=total_cases
```

```python
new_cases=death['new_cases'].values
df['new_cases']=new_cases
```

```python
Percent_of_death=np.zeros((173,1),dtype='float16')
new_cases=df['new_cases'].values.astype(float)
new_death=df['New_death'].values.astype('uint16')
for i in range(173):
    Percent_of_death[i]=(new_death[i]/new_cases[i])*100

Percent_of_death=Percent_of_death.astype('float16')    
```

```python
df['Percent_of_death']=Percent_of_death
```

```python
Percentage_of_virus=np.zeros((173,1),dtype='float16')
total_cases=df['total_cases'].values.astype(float)
new_cases=df['new_cases'].values.astype(float)
for i in range(173):
    Percentage_of_virus[i]=(new_cases[i]/total_cases[i])*100
```

```python
df['Percentage_of_virus']=Percentage_of_virus
```

```python
df
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cbar=False)
plt.tight_layout(pad=0.5)
plt.show()
```

## Scaling & split data

Predict the number of injected vaccines to achieve less death and favorable

```python
from sklearn.model_selection import train_test_split
x=df.iloc[:,[0,5]]
y=df.iloc[:,3]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Train your model

```python
from sklearn.linear_model import LinearRegression
linear_Reg=LinearRegression()
linear_Reg.fit(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.1)
ridge.fit(x_train,y_train)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear',C=1.0)
svr.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest=RandomForestRegressor()
random_forest.fit(x_train,y_train)
```

```python
import xgboost
```

```python
from xgboost import XGBRegressor
xg=XGBRegressor(n_estimators=500)
xg.fit(x_train,y_train)
```

## Test the model and show the metrics

```python
y_linear=linear_Reg.predict(x_test).astype('uint32')
```

```python
y_ridge=ridge.predict(x_test).astype('uint32')
```

```python
y_svr=svr.predict(x_test).astype('uint32')
```

```python
y_tree=tree.predict(x_test).astype('uint32')
```

```python
y_random_forest=random_forest.predict(x_test).astype('uint32')
```

```python
y_xg=xg.predict(x_test).astype('uint32')
y_xg
```

```python
y_pred=[y_linear,y_ridge,y_svr,y_tree,y_random_forest,y_xg]
```

```python
from sklearn.metrics import r2_score
r2_score_=[]
for y in y_pred:
    r2=r2_score(y_test,y)
    r2_score_.append(r2)
```

```python
r2_score_
```
